```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv("George Lifting Log - Sheet1.csv")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
```

```python
plt.plot(df["Date"], df["Epley_1RM"])
```

```python
df_squat = df[df["Lift"] == "Squat"]
plt.plot(df_squat["Date"], df_squat["Epley_1RM"])
plt.plot(df_squat["Date"], df_squat["Weight"])
plt.xticks(rotation=40)
```

```python
df_deadlift = df[df["Lift"] == "Deadlift"]
plt.plot(df_deadlift["Date"], df_deadlift["Epley_1RM"])
plt.plot(df_deadlift["Date"], df_deadlift["Bryzcki_1RM"])
plt.plot(df_deadlift["Date"], df_deadlift["Weight"])
plt.xticks(rotation=40)
```

```python
df_all_deadlift = df[(df["Lift"] == "Deadlift") | (df["Lift"] == "Hex Bar Deadlift")]
plt.plot(df_all_deadlift["Date"], df_all_deadlift["Epley_1RM"])
plt.plot(df_all_deadlift["Date"], df_all_deadlift["Weight"])
plt.xticks(rotation=40)
```

```python
plt.plot(df_deadlift["Date"], df_deadlift["Epley_1RM"])
plt.plot(df[df["Lift"] == "Hex Bar Deadlift"]["Date"], df[df["Lift"] == "Hex Bar Deadlift"]["Epley_1RM"])
plt.xticks(rotation=40)
```

```python
df_bench = df[df["Lift"] == "Bench Press"]
plt.plot(df_bench["Date"], df_bench["Epley_1RM"])
plt.xticks(rotation=40)
```

```python
df_ohp = df[(df["Lift"] == "Overhead Press") | (df["Lift"] == "Push Press")]
plt.plot(df_ohp["Date"], df_ohp["Epley_1RM"])
plt.xticks(rotation=40)
```
